```python
#Load necessaries libraries
from fastai import *
from fastai.vision import *
from torchvision.models import *
import numpy as np
from PIL import Image
```

```python
#Initialize architecture
arch = resnet50   # specify model architecture.
bs = 8   # specify batch size
sz = 512   # input size is the crop size
```

```python
#Initialize folder structure for architecture, this is requuired prior to loading the
weights
PATH ='Path to folder where train and val folder are located'
tfms = get_transforms(do_flip=True,max_zoom=1.3)
data = ImageDataBunch.from_folder('./',ds_tfms = tfms,valid='test', size=sz,bs=bs)
learner = cnn_learner(data, arch, pretrained=True, path='.', metrics=accuracy, ps=0.2,
callback_fns=ShowGraph)
```

```python
#Load saved weights during training
learner.load('Path to saved weights')
```

```python
#Open image for inference
img = open_image('Path to image for inference purpose')
```

```python
#Make prediction
prediction = to_np(learn.predict(img)[2]).argmax()
```

```python
#Model Output
if (prediction == 0):
    print('No-DR')
elif (prediction == 1):
    print('Mild-DR')
elif (prediction == 2):
    print('Moderate-DR')
elif (prediction == 3):
    print('Severe-DR')
else:
    print('Proliferative-DR')
```
