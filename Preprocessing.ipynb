```python
import cv2
import os
import numpy as np
```

```python
def coord(img):     #Find coordinates of center patch to be cropped
    top = 0         #Excluding the black perimeter
    for i in range(img.shape[0]):
        if sum(img[i,:]) == 0:
            continue
        top = i
        break

    botm = img.shape[0]
    for i in range(img.shape[0]):
        if sum(img[-i-1,:]) == 0:
            continue
        botm -= i
        break

    left = 0
    for i in range(img.shape[1]):
        if sum(img[:,i]) == 0:
            continue
        left = i
        break

    right = img.shape[1]
    for i in range(img.shape[1]):
        if sum(img[:,-i-1]) == 0:
            continue
        right -= i
        break
    
    return [top,botm,left,right]
```

```python
def crop_resize(path):
    img = cv2.imread(path) #Load original image
    temp = cv2.cvtColor(img,cv2.COLOR_RGB2GRAY) #Convert to gray
    temp[temp <= 13] = 0 #Remove noise
    dim = coord(temp) #Get points of center image
    imgnew = img[dim[0]:dim[1],dim[2]:dim[3]] # Cropping the center part
    imgnew = cv2.resize(imgnew,(512,512)) #Resize the image to 512x512
    cv2.imwrite(path,imgnew) #Save the image
```

```python
import os
from sklearn import io, exposure
```

```python
def data_aug_val(path):
    listing_test = os.listdir(path)
    for file_name in listing_test:
        image = io.imread(path+file_name) #Read Image
        sigmoid_correction_image = exposure.adjust_sigmoid(image) #Apply correction
        io.imsave(path+'aug-3'+file_name,sigmoid_correction_image) #Save image
```
