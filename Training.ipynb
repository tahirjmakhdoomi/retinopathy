```python
from fastai import *
from fastai.vision import *
from torchvision.models import *   # import necessary libraries
```

```python
arch = resnet50   # specify model architecture
bs = 8     # specify batch size
sz = 512   # input size is the crop size
```

```python
PATH ='Path to root of dataset folder containing train and val folder'
tfms = get_transforms(do_flip=True,max_zoom=1.3)
data = ImageDataBunch.from_folder(PATH,ds_tfms=tfms, valid='val', size=sz,bs=bs) #load data
data.normalize(data.batch_stats()) #Normalize data
```

```python
#Load CNN model
learner = cnn_learner(data, arch, pretrained=True, path='.', metrics=accuracy, ps=0.5,
callback_fns=ShowGraph)
```

```python
iter_count = 600 #set no of iterations
learner.lr_find(wd=1e-6, num_it=iter_count) #Set weight decay and start finding the LR
learner.recorder.plot(show_grid=True) #Plot LR graph
```

```python
# Training the Model
```

```python
arch = resnet50   # specify model architecture
bs = 8     # specify batch size
sz = 512   # input size is the crop size
```

```python
PATH ='Path to root of dataset folder containing train and val folder'
tfms = get_transforms(do_flip=True,max_zoom=1.3)
data = ImageDataBunch.from_folder(PATH,ds_tfms=tfms, valid='val', size=sz,bs=bs) #load data
data.normalize(data.batch_stats()) #Normalize data
```

```python
#Load CNN model
learner = cnn_learner(data, arch, pretrained=True, path='.', metrics=accuracy, ps=0.5,
callback_fns=ShowGraph)
```

```python
iter_count = 600 #set no of iterations
cyc_len = 8 #Set no of epochs
max_lr = 1e-3 #Set base LR
wd = 1e-6 #set weight decay
learner.fit_one_cycle(cyc_len, max_lr, wd) #Start training the model
learner.save('resnet50') #Save model weights
```
